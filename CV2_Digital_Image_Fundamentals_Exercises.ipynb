Image digitization involves two steps:
1. **Sampling**: Dividing the image into a grid of pixels.
2. **Quantization**: Assigning discrete intensity values to each pixel.

### Task:
1. Create a continuous grayscale image using a sine wave function.
2. Perform sampling to reduce the resolution to 10x10.
3. Perform quantization by rounding values to the nearest integer in the range [0, 255].
4. Display the original and processed images side by side.

### Hint:
- Use `np.linspace` for creating the grid.
- Use `np.round` for quantization.


```python

# Generate continuous grayscale image (sine wave pattern)
x = np.linspace(0, 2 * np.pi, 100)
y = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)
continuous_image = (np.sin(X) + np.sin(Y)) * 127.5 + 127.5

# Sampling: Reduce resolution to 10x10
sampled_image = continuous_image[::10, ::10]

# Quantization: Round values to nearest integer in range [0, 255]
quantized_image = np.round(sampled_image)

# Display original and processed images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(continuous_image, cmap='gray', vmin=0, vmax=255)
axs[0].set_title("Original Image")
axs[1].imshow(quantized_image, cmap='gray', vmin=0, vmax=255)
axs[1].set_title("Sampled & Quantized Image")
plt.show()
```

# **Introduction to Intensity Level Slicing**

**Intensity Level Slicing** is an image enhancement technique used to highlight or preserve specific intensity ranges in an image. This technique is particularly useful for tasks such as:

- **Highlighting features** within a specific intensity range (e.g., detecting tumors in medical images).
- **Suppressing unwanted regions** or intensities outside the target range.
- **Isolating objects** based on intensity values.

### **How It Works**
For a given intensity range \([A, B]\):
1. **Highlight intensities in \([A, B]\):**
   All pixels with intensity values within the range are set to a high value (e.g., 255 for 8-bit images), while others are suppressed.

2. **Preserve intensities in \([A, B]\):**
   Pixels within the range are kept unchanged, while those outside are suppressed to a low value (e.g., 0).

### **Applications**
- **Medical Imaging:** Highlighting specific intensity ranges to detect abnormalities.
- **Satellite Imaging:** Isolating features such as water bodies or vegetation.
- **Industrial Imaging:** Detecting defects in materials or products.

# **Notes on Sampling and Quantization**

### **1. Introduction**
When converting a continuous image (analog) to a digital image, two key processes are involved:
1. **Sampling:** Dividing the image into a discrete grid of pixels.
2. **Quantization:** Assigning discrete intensity values to each pixel.

Both steps are critical for digital image representation and compression.

---

### **2. Sampling**
**Definition:** Sampling determines the spatial resolution of an image by breaking it into a grid of pixels.

- **High Sampling Rate:** More pixels, higher resolution, better detail.
- **Low Sampling Rate:** Fewer pixels, lower resolution, possible loss of detail (aliasing).

#### **How It Works:**
- Imagine an image as a continuous surface.
- Sampling involves selecting points (pixels) at regular intervals to approximate the original surface.

#### **Example:**
Reducing an image's resolution from \(100 \times 100\) to \(10 \times 10\) is a form of sampling.

---

### **3. Quantization**
**Definition:** Quantization maps the continuous range of intensity values in the image to a finite set of discrete levels.

- **High Quantization Levels:** More intensity levels, smoother transitions.
- **Low Quantization Levels:** Fewer intensity levels, visible banding.

#### **How It Works:**
- The continuous intensity range (e.g., 0–255 for 8-bit images) is divided into discrete intervals.
- Each pixel intensity is rounded to the nearest interval value.

#### **Example:**
Mapping grayscale values to 4 levels instead of 256 reduces detail but saves storage space.

---

### **4. Combined Impact**
Both sampling and quantization affect the quality of a digital image:
- **Low Sampling Rate:** Loss of spatial resolution, blurry images.
- **Low Quantization Levels:** Loss of intensity resolution, banding effects.

---

### **5. Applications**
- **Image Compression:** Reducing file size (e.g., JPEG compression).
- **Image Processing:** Standardizing images for analysis.
- **Medical Imaging:** Efficient storage of high-resolution scans.

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a continuous image (e.g., a sine wave pattern)
x = np.linspace(0, 2 * np.pi, 100)
y = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)
continuous_image = np.sin(X) * np.sin(Y) * 127.5 + 127.5

# Sampling: Reduce resolution to 10x10
sampled_image = continuous_image[::10, ::10]

# Display original and sampled images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(continuous_image, cmap='gray', vmin=0, vmax=255)
axs[0].set_title("Original Image (100x100)")
axs[1].imshow(sampled_image, cmap='gray', vmin=0, vmax=255)
axs[1].set_title("Sampled Image (10x10)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a grayscale image with values from 0 to 255
grayscale_image = np.random.randint(0, 256, size=(100, 100), dtype=np.uint8)

# Perform quantization (reduce to 8 levels)
num_levels = 8
quantized_image = (grayscale_image // (256 // num_levels)) * (256 // num_levels)

# Display original and quantized images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(grayscale_image, cmap='gray', vmin=0, vmax=255)
axs[0].set_title("Original Image (256 Levels)")
axs[1].imshow(quantized_image, cmap='gray', vmin=0, vmax=255)
axs[1].set_title(f"Quantized Image ({num_levels} Levels)")
plt.show()
```

```python
# Step 1: Create a continuous grayscale image (sine wave pattern)
x = np.linspace(0, 2 * np.pi, 100)  # High resolution
y = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)
continuous_image = (np.sin(X) + np.sin(Y)) * 127.5 + 127.5

# Step 2: Reduce resolution to simulate sampling
sampled_image = continuous_image[::10, ::10]

# Step 3: Quantize to 4 levels
num_levels = 4
quantized_sampled_image = (sampled_image // (256 // num_levels)) * (256 // num_levels)

# Step 4: Display all images
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(continuous_image, cmap='gray', vmin=0, vmax=255)
axs[0].set_title("Original Image (100x100)")
axs[1].imshow(sampled_image, cmap='gray', vmin=0, vmax=255)
axs[1].set_title("Sampled Image (10x10)")
axs[2].imshow(quantized_sampled_image, cmap='gray', vmin=0, vmax=255)
axs[2].set_title("Quantized Sampled Image (4 Levels)")
plt.show()
```

```python
# Step 1: Create a grayscale image (sine wave)
x = np.linspace(0, 2 * np.pi, 100)  # High resolution
y = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)
grayscale_image = (np.sin(X) + np.sin(Y)) * 127.5 + 127.5

# Step 2: Sample at different rates
# Replace '...' with appropriate slicing to reduce resolution
sampled_image_1 = grayscale_image[::5, ::5]  # Reduce to 20x20
sampled_image_2 = grayscale_image[::10, ::10]  # Reduce to 10x10

# Step 3: Display the images
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(grayscale_image, cmap='gray', vmin=0, vmax=255)
axs[0].set_title("Original Image (100x100)")
axs[1].imshow(sampled_image_1, cmap='gray', vmin=0, vmax=255)
axs[1].set_title("Sampled Image (20x20)")
axs[2].imshow(sampled_image_2, cmap='gray', vmin=0, vmax=255)
axs[2].set_title("Sampled Image (10x10)")
plt.show()
```
